Projeto 2: Criar um periodo de onda quadrada com soma de cossenos

Parte 1: Contrução da onda quadrada:

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = 10
def sq(x, t):
    square=0
    for i in range(1, x+1):
        square = square + ((np.sin(((i*2)-1)*t))/((2*i)-1))
        
    return(square)
t = np.arange(0.0, 10 ,0.001)


plt.plot(t,sq(x,t))
```
