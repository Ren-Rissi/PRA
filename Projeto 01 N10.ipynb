```python
import matplotlib.pyplot as plt
import numpy as np
import random as rd
%matplotlib inline

def f(h, a0, a1, f1, f2, f3):
    if h>=f3:
         return(f(h-f3, a0, a1, f1, f2, f3))
    else:
        g = h*a0
        ff1 = np.cos(f1*np.pi*h)*0.3
        ff2 = np.cos(f2*np.pi*h)
        return(g+ff1+ff2)

def sh():
    t = np.arange(0.0,10,0.001)
    a0 = rd.randint(-5,5)
    a1 = rd.randint(-5,5)
    f1 = rd.randint(3,6)
    f2 = rd.randint(0,3)
    f3 = rd.random()*10
    x = np.arange(0.0,10,0.001)
    for i in range(0,len(x)):
        x[i] = f(t[i],a0,a1,f1,f2,f3)
    plt.plot(t,x)
    return
sh()
```
