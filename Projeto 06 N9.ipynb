Projeto 6: Encontrar conjuntos de dados e mostrar corr, cov e corrP;
mínimo 30 individuos, e 4 medidas pra cada indvíduo. Apresentar como uma matriz de gráficos

```python
import numpy as np
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats.stats import pearsonr 


#investimento em agricultura em porcentagem do pib de 1995 até 2015
#Brasil=1,China=2,Argentina=3,Mundo=4

data = np.array([[5.8,5.5,5.3,5.4,5.4,5.5,5.6,6.4,7.2,6.7,5.5,5.1,5.2,5.4,5.2,4.8,5.1,4.9,5.3,5.0,5.0],
[19.6,19.3,17.9,17.2,16.1,14.7,14.0,13.3,12.3,12.9,11.6,10.6,10.3,10.3,9.8,9.5,9.4,9.4,9.3,9.1,8.8],
[5.8,6.1,5.7,5.7,4.8,5.1,4.9,10.8,11.0,9.8,9.3,8.2,8.8,8.7,6.3,8.5,8.3,6.9,7.2,8.0,6.0],
[8.0,7.9,6.0,5.9,5.6,5.2,5.1,5.0,4.9,4.8,4.4,4.1,4.0,4.0,3.9,3.9,4.0,3.9,4.0,3.9,3.8]])
data_cov=np.cov(data)
data_corr=np.corrcoef(data)

fig = plt.figure(1,figsize=(13.8,13.8))
gridspec.GridSpec(4,4)
for i in range(4):
    for j in range(4):
        plt.subplot2grid((4,4), (i,j))
        plt.title('Graf '+str(i+1)+"-"+str(j+1))
        plt.text(0.9*abs(max(data[i,:])),0.9*abs(max(data[j,:])),'Cov: '+str("{0:.4f}".format(data_cov[i,j]))+
                 '\nCorr: '+str("{0:.4f}".format(data_corr[i,j]))+'\nCorP:'+
                 str("{0:.4f}".format(pearsonr(data[:,i],data[:,j])[0])),bbox=dict(facecolor='white', alpha=0.3))
        a=plt.plot(data[i,:],data[j,:],'.')
        plt.tight_layout()
```
